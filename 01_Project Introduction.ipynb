## 1. Project Overview
The goal of this project is to predict the quality of wine based on its physicochemical properties.

We use machine learning models to analyze patterns in the data and estimate wine quality scores.

This notebook serves as an introduction and overview of the full project pipeline.


## 2. Dataset Description
The dataset contains physicochemical properties of wine samples from the Portuguese *Vinho Verde* wine region. It includes both red and white wine variants.

Due to privacy and logistical constraints, only:
- Physicochemical features (inputs)
- Sensory quality scores (output)

are available. Information such as grape variety, brand, or price is not included.

### 2.1 Features
All variables are continuous numerical features with no missing values. The main features include:
- fixed_acidity
- volatile_acidity
- citric_acid
- residual_sugar
- chlorides
- free_sulfur_dioxide
- total_sulfur_dioxide
- density
- pH
- sulphates

These features describe the chemical composition of the wine and are used to predict its quality.

### 2.2 Target Variable
- quality: a numerical score representing wine quality


Remark: Important Characteristics
- The problem can be treated as regression or classification
- The quality labels are ordered but imbalanced (most wines are of average quality)
- Extreme values (very high or very low quality wines) are rare and may be considered outliers
- Not all input variables may be equally relevant, making feature selection an interesting extension

This makes the dataset suitable for testing different machine learning approaches and evaluating model performance under realistic conditions.

### 2.3 Evaluation Metrics
We evaluate models using:
- Mean Squared Error (MSE)
- R² Score


These metrics help assess prediction accuracy and model performance.
